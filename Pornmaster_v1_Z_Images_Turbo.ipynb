```python
import torch
from diffusers import StableDiffusionXLPipeline

# 1) Set the model ID from Hugging Face
model_id = "UnfilteredAI/NSFW-gen-v2"

# 2) Load the pipeline (text-to-image)
pipe = StableDiffusionXLPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16,  # use fp16 if GPU supports for better speed
    use_safetensors=True
)

# Send model to GPU
pipe.to("cuda")

# 3) Your text prompt
prompt = "A detailed, uncensored (NSFW) fantasy portrait"

# Optionally add a negative prompt
negative_prompt = "blurry, low quality, ugly"

# 4) Generate image
result = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt,
    guidance_scale=7.5,
    num_inference_steps=30,
    width=1024,
    height=1024,
)

# 5) Save the result
image = result.images[0]
image.save("nsfw_output.png")

print("Image saved as nsfw_output.png")
```
